### 2. Write a program to implement simple linear regression for the given data (Both) 

| Weight (lbs.) | Height |
|---------------|--------|
| 140           | 60     |
| 155           | 62     |
| 159           | 67     |
| 179           | 70     |
| 192           | 71     |
| 200           | 72     |
| 212           | 75     |

#### From Scratch

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Datasets/weightvsheight.csv")
x = data["Weight"]
y = data["Height"]
n = len(x)

xMean = sum(x) / n
yMean = sum(y) / n

dx = x - xMean
dy = y - yMean

d = dx * dy
dx2 = dx * dx
dy2 = dy * dy

slope = sum(d) / sum(dx2)
intercept = yMean - slope * xMean

yp = slope * x + intercept

r2 = 1 - (sum((y - yp) ** 2) / sum(dy2))

plt.scatter(x, y, c='b')
plt.plot(x, yp, c='r')

print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
print(f'R^2: {r2}')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend(['Data Points', f'Best Fit: y = {slope:.2f}x + {intercept:.2f}'])
plt.title('Simple Linear Regression')
plt.grid(True)
plt.show()
```

#### Using Library

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv("Datasets/weightvsheight.csv")
x = data[["Weight"]]
y = data["Height"]

model = LinearRegression()
model.fit(x, y)

yp = model.predict(x)

slope = model.coef_[0]
intercept = model.intercept_

plt.scatter(x, y, c='b')
plt.plot(x, yp, c='r')
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend(['Data Points', f'Best Fit: y = {slope:.2f}x + {intercept:.2f}'])
plt.title('Simple Linear Regression')
plt.grid(True)
plt.show()
```
