## 3. Given two empty jugs of quantity m and n litres respectively. The jugs don't have any measurement markings to allow measuring smaller quantities. The task is to find the minimum number of operations to be performed to obtain d litres of water in m litres jug Write a program to solve water jug problem using BFS and DFS technique.

```python
from collections import deque
import math
import sys

sys.setrecursionlimit(2000)

def water_jug_bfs(m, n, d):
    if d > max(m, n) or d % math.gcd(m, n) != 0:
        return []

    start_state = (0, 0)
    
    if d == 0:
        return [start_state]

    queue = deque([(start_state, [start_state])])
    visited = {start_state}

    while queue:
        current_state, path = queue.popleft()
        x, y = current_state

        next_states = [
            (m, y),
            (x, n),
            (0, y),
            (x, 0),
            (x - min(x, n - y), y + min(x, n - y)),
            (x + min(y, m - x), y - min(y, m - x))
        ]

        for state in next_states:
            if state not in visited:
                new_path = path + [state]
                if state[0] == d or state[1] == d:
                    if state[1] == d and state[0] != d:
                         if state[0] + state[1] <= m:
                            new_path.append((state[0] + state[1], 0))
                    return new_path
                
                visited.add(state)
                queue.append((state, new_path))
    return []

def water_jug_dfs_recursive(m, n, d):
    if d > max(m, n) or d % math.gcd(m, n) != 0:
        return []
    visited = set()
    def dfs_helper(path):
        current_state = path[-1]
        if current_state in visited:
            return None
        
        visited.add(current_state)
        
        x, y = current_state
        if x == d or y == d:
            final_path = list(path)
            if y == d and x != d:
                if x + y <= m:
                     final_path.append((x + y, 0))
            return final_path

        next_states = [
            (m, y),
            (x, n),
            (0, y),
            (x, 0),
            (x - min(x, n - y), y + min(x, n - y)),
            (x + min(y, m - x), y - min(y, m - x))
        ]

        for state in next_states:
            solution_path = dfs_helper(path + [state])
            if solution_path:
                return solution_path
        
        return None

    return dfs_helper([(0, 0)]) or []

if __name__ == '__main__':
    m_capacity = 4
    n_capacity = 3
    desired_amount = 2

    print("BFS:")
    path_bfs = water_jug_bfs(m_capacity, n_capacity, desired_amount)
    if path_bfs:
        for state in path_bfs:
            print(state)
            
    print("\nDFS:")
    path_dfs = water_jug_dfs_recursive(m_capacity, n_capacity, desired_amount)
    if path_dfs:
        for state in path_dfs:
            print(state)
```
