## 6. Write a program to implement simple linear regression

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'X': [23, 35, 2, 28, 9, 32, 25, 10],
    'Y': [486, 747, 81, 581, 112, 728, 445, 128]
})

x = data['X']
y = data['Y']
n = len(x)

xMean = sum(x) / n
yMean = sum(y) / n

dx = x - xMean
dy = y - yMean

d = dx * dy
dx2 = dx * dx
dy2 = dy * dy

slope = sum(d) / sum(dx2)
intercept = yMean - slope * xMean

yp = slope * x + intercept

r2 = 1 - (sum((y - yp) ** 2) / sum(dy2))

plt.scatter(x, y, c='b')
plt.plot(x, yp, c='r')

print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
print(f'R^2: {r2}')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.legend(['Data Points', f'Best Fit: y = {slope:.2f}x + {intercept:.2f}'])
plt.title('Simple Linear Regression')
plt.grid(True)
plt.show()
```
