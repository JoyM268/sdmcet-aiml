```python
def print_board(board): 
    for i in range(6): 
        for j in range(6): 
            print(board[i][j] if board[i][j] != 0 else ".", end=" ")

            if (j + 1) % 3 == 0 and j != 5: 
                print("|", end=" ") 
        print() 
        if (i + 1) % 2 == 0 and i != 5: 
            print("-" * 21) 
 
def find_empty(board): 
    for i in range(6): 
        for j in range(6): 
            if board[i][j] == 0: 
                return (i, j) 
    return None 
 
def is_valid(board, row, col, num): 
    if num in board[row]: 
        return False 
 
    for i in range(6): 
        if board[i][col] == num: 
            return False 
 
    start_row = (row // 2) * 2 
    start_col = (col // 3) * 3 
    for i in range(2): 
        for j in range(3): 
            if board[start_row + i][start_col + j] == num: 
                return False 
 
    return True 
 
def solve_sudoku(board):  
    empty = find_empty(board) 
    if not empty: 
        return True 
    row, col = empty 
 
    for num in range(1, 7): 
        if is_valid(board, row, col, num): 
            board[row][col] = num 
            if solve_sudoku(board): 
                return True 
            board[row][col] = 0
 
    return False 


sudoku_board = [ 
    [0, 0, 0, 0, 6, 0], 
    [0, 0, 0, 0, 0, 4], 
    [0, 0, 2, 0, 0, 0], 
    [6, 0, 0, 0, 0, 0], 
    [2, 0, 0, 3, 0, 0], 
    [0, 1, 0, 0, 0, 0] 
] 
 
print("Initial 6x6 Sudoku Puzzle:") 
print_board(sudoku_board) 
 
if solve_sudoku(sudoku_board): 
    print("\nSolved Sudoku:") 
    print_board(sudoku_board) 
else: 
    print("No solution exists.") 
```
